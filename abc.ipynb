# Day of Birth predictor

```python
print ("**********************Use ""/"" only in date format**************************")

Name=input ("Enter your Name:- ")

print ("*********************------------------------******************************\n")

#Dictionaries of days,non leap years days, leap years days.
days={1:"Monday",2:"Tuesday",3:"Wednesday",4:"Thursday",5:"Friday",6:"Saturday",0:"Sunday"}
Num_non={1:31,2:28,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}
Num_Leap={1:31,2:29,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}

#No. of odd days in leap year of 100th,200th,300th and 400th year
year_100=((76*1)+(24*2))%7
year_200=(year_100*2)%7
year_300=(year_100*3)%7
year_400=0



print ("**************Hello {} , Thank you for being here**************************\n".format(Name))
print("\n")

date=input("enter your Birth date :- dd/mm/yyyy ")

dob=date.split("/")
for i in range(0,len(dob)):
    dob[i]=int(dob[i])


if dob[2]%4==0:
    total_days=366
    odd_day=2
    Num_Leap
    nd=dob[2]-1
    cal=nd-400*(nd//400)
    val=100*(cal//100)
    if val==cal:
          val=0
    leap_years_=(nd-(nd-cal+val))//4
    Non_leap_years=(nd-(nd-cal+val))-leap_years_
    if val == 100:
        odd=year_100
        
    elif val == 200:
        odd=year_200
        
    elif val == 300:
        odd=year_300
    
    else:
        odd=year_400
        
    Num_days_BirthYear=0
    for i in range(1,dob[1]):
         Num_days_BirthYear+=Num_Leap[i]
    
    Total_odd=(odd+(leap_years_*2)+Non_leap_years+Num_days_BirthYear+dob[0])%7
    
    print("-------------------------------******---------------------------------\n")
    print (Name,"the day you are born is ",days[Total_odd])
    print ("\n")
    print("-------------------------------******---------------------------------\n")
else:
    total_days=365
    odd_day=1
    Num_non
    nd=dob[2]-1
    cal=nd-400*(nd//400)
    val=100*(cal//100)
    if val==cal:
        val=0
    leap_years_=(nd-(nd-cal+val))//4
    Non_leap_years=(nd-(nd-cal+val))-leap_years_
    if val == 100:
         odd=year_100
        
    elif val == 200:
        odd=year_200
        
    elif val == 300:
        odd=year_300
    
    else:
        odd=year_400
        
    Num_days_BirthYear=0
    for i in range(1,dob[1]):
        Num_days_BirthYear+=Num_non[i]
    
    Total_odd=(odd+(leap_years_*2)+Non_leap_years+Num_days_BirthYear+dob[0])%7
    print("-------------------------------******---------------------------------\n")
    print (Name,"the day you are born is ",days[Total_odd])
    print ("\n")
    print("-------------------------------******---------------------------------\n")

    
   
import datetime
today=datetime.datetime.today()
to=str(today).split()[0].split('-')

for i in to:
    if int(to[2])>dob[0]:
        day=int(to[2])-dob[0]
    else:
        day=dob[0]-int(to[2])
    if int(to[1])>dob[1]:
        month=12-(int(to[1])-dob[1])
    else:
        month=12-(dob[1]-int(to[1]))
    year=int(to[0])-dob[2]

print ("************************----------------------******************************")    
    
print ("You are on earth since ",year, "years ",month, " months","and ", day, " days")

print("*************************-----------------------******************************")
```
